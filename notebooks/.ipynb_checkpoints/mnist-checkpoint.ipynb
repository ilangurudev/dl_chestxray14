```python
import fastai
from fastai.vision import *
```

```python
mnist = untar_data(URLs.MNIST_SAMPLE)
```

```python
data = ImageDataBunch.from_folder(mnist, ds_tfms=get_transforms()).normalize(imagenet_stats)
model = create_cnn(data, models.resnet18, metrics=[accuracy])
model.fit_one_cycle(5)
```

```python
data.export()
model.save('model')
```
